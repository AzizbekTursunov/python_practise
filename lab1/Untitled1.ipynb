```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.01)
plt.figure(figsize = (15, 7))
plt.plot(x, np.sin(x), label = (r' $f_1(x) = \sin(x) $'))
plt.plot(x, np.cos(x), label = (r'$ f_2(x)  = \cos(x) $'))
plt.plot(x, -x, label = (r'$ f_3(x)  = -x $'))
plt.xlabel(r'$ x $', fontsize = 24)
plt.ylabel(r'$ f(x) $', fontsize = 24)
plt.legend(loc = 'best', fontsize = 24)
plt.title(r'$ uch - funksiya - grafigi $')
plt.grid(True)
plt.savefig('sin(x)_cos(x)_-x_grafigi.png')
plt.show()
```
